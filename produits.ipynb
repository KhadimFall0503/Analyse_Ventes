# Liste des produits les plus vendus par categorie

```python
 liste_produits = df.groupby(["produit", "categorie"])["quantite"] \
  .sum() \
  .reset_index() \
  .sort_values("quantite", ascending=False)
```

```python
liste_produits
```

# Impoter la librairie matplotlib et Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = liste_produits["produit"]
y = liste_produits["quantite"]
```

# Liste des Produits les plus vendus par Categories

```python


fig, ax = plt.subplots(figsize=(12,8))
bars = ax.bar(x, y, label="Quantité vendue",color="#4E79A7")

ax.set_title("Liste des Produits les plus vendus par Catégories", fontname="Arial", fontweight="bold", fontsize=19)
ax.set_xlabel("Liste des Produits", fontname="Arial", fontweight="bold", fontsize=19)
ax.set_ylabel("Nombre de Produits vendus", fontname="Arial", fontweight="bold", fontsize=19)
ax.legend()
ax.set_xticklabels(x, rotation=45)
ax.grid(False)

# ➕ Afficher les valeurs automatiquement
ax.bar_label(bars, fmt='%.0f', fontsize=12, fontweight='bold')

plt.show()
```

# Répartition des ventes par produit (trié par date de vente) 

```python
couleur = [
    "#4E79A7",  # Bleu doux
    "#F28E2B",  # Orange chaud
    "#E15759",  # Rouge corail
    "#76B7B2",  # Vert-bleu pastel
    "#59A14F",  # Vert nature
    "#EDC948",  # Jaune doré
    "#B07AA1",  # Violet lavande
    "#FF9DA7",  # Rose clair
    "#9C755F",  # Marron clair
    "#BAB0AC"   # Gris neutre
]

plt.figure(figsize=(8,8))
plt.pie(df['quantite'], 
        labels=df['produit'], 
        autopct='%1.1f%%', 
        startangle=100, 
        shadow=False, 
        colors= couleur)

plt.title("Répartition des ventes par produit (trié par date de vente)",
          fontname="Arial", fontweight="bold", fontsize=16)
plt.show()
```
