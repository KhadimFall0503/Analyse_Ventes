# Impoter la librairie pandas

```python
import pandas as pd
```

# Transformer le fichier ventes.txt en ventes.csv

```python
df = pd.read_csv("ventes.txt",sep= ",")
df.to_csv("ventes.csv", index=False)
```

```python
df.sort_values("prix_unitaire")
```

# Crée une colonne montant_total = quantite * prix_unitaire.

```python
df["montant_total"] = df["quantite"] * df["prix_unitaire"]
```

```python
df.head()
```

# Trouve le produit qui a le plus grand montant total.

```python
df.groupby("produit")["montant_total"].max()
```

 # Affiche uniquement les ventes dont le prix unitaire > 100 €.

# Trouve la quantité moyenne vendue par catégorie.

```python
df.groupby("categorie")["quantite"].mean()
```

# Convertis date_vente en format datetime (pd.to_datetime).

```python
df["date_vente"] = pd.to_datetime(df["date_vente"])
```
